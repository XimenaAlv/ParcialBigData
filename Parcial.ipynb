```python
URL = "https://es.wikipedia.org/wiki/Anexo:Presidentes_de_Colombia"
```

```python
page = requests.get(URL)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
tables = soup.find_all("table", class_="wikitable")

len(tables)
```

```python
presidents=[]

for table in tables:            
    for row in table.find_all(['tr']):
        
        attributes = row.find_all(['td'])        
        if attributes == []: continue
        
        if len(attributes) == 7: continue
            
        presidents.append({
            "Período": attributes[1].getText().replace("\n",""),
            #"Foto": attributes[2].find('img')['src'],
            "Nombre": attributes[3].getText().replace("\n",""),
            "Partido": attributes[4].getText().replace("\n",""),            
        })

presidents
```
